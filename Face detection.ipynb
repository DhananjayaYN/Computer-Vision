```python
import cv2
```

```python
haar_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") #Add defult file
```

```python
cam =cv2.VideoCapture(0)

while True:
    _,img = cam.read()
    grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    faces = haar_cascade.detectMultiScale(grayImg,1.3,4) 
    for(x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2) # create rectangle around the face

    cv2.imshow("Face Detection",img)

    key = cv2.waitKey(5)
    print(key)
    if key == 27 :#esc key
        break

cam.release()
cv2.destroyAllWindows()
    
```
