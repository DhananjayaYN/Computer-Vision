```python
from facial_emotion_recognition import EmotionRecognition as em
import cv2
```

```python
emotion = em(device='cpu')
cam = cv2.VideoCapture(0)

while True:
    success,frame = cam.read()
    frame = emotion.recognise_emotion(frame , return_type= 'BGR')
    cv2.imshow('Frame' , frame)
    key = cv2.waitKey(10)
    if key == 27:
        break

cam.release()
cv2.destroyAllWindows()
    
```
