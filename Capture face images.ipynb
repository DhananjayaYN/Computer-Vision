```python
import cv2
import os
```

```python
haar_file ='haarcascade_frontalface_default.xml'
datasets = 'datasets'
sub_data = 'Steve Harvey'
```

```python
path = os.path.join(datasets, sub_data) #create the path to store the images
if not os.path.isdir(path) :
    os.mkdir(path)
```

```python
(width , height) = (130, 100) # Set the width and height for resizing the detected face images

face_cascade = cv2.CascadeClassifier(haar_file)

webcam = cv2.VideoCapture(0)
count = 0
while count < 101 :
    print(count)
    (_,img) = webcam.read()
    grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(grayImg , 1.3 , 4) # Detect faces in the grayscale image using the face cascade classifier

    for (x,y,w,h) in faces:
        cv2.rectangle(img , (x,y), (x+w, y+h), (0,255,0), 2)
        face = grayImg[y:y+h , x:x+w] # Extract the face region from the grayscale image
        face_resize = cv2.resize(face, (width,height))
        cv2.imwrite('%s/%s.png'%(path,count), face_resize)

    count +=1

    cv2.imshow("Frame",img)
    key = cv2.waitKey(10)
    if key == 27:
        break

webcam.release()
cv2.destroyAllWindows()
    
```
