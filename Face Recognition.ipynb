```python
import cv2
import numpy as np
import os
```

```python
harr_file = 'haarcascade_frontalface_default.xml'
datasets = 'datasets'
```

```python
(images , labels , names , id ) = ([],[],{},0) 
print("Training...")

# Traverse through all directories and files in the dataset folder
for (subdirs , dirs , files) in os.walk(datasets):
    for subdir in dirs:
        names[id] = subdir # Map the current ID to the subdirectory name (person's name)
        subjectpath = os.path.join(datasets , subdir)  # Create the full path to the subdirectory

        # Loop through each file in the subject's directory
        for filename in os.listdir(subjectpath):
            path = subjectpath + '/' + filename  # Get the full path of the image
            label = id
            images.append(cv2.imread(path,0))
            labels.append(int(label))

        id +=1

(width , height) = (130, 100)

# Convert the images and labels lists into NumPy arrays for training
(images , labels) = [np.array(lis) for lis in [images , labels]]

# Initialize the FisherFace Recognizer model for training
model = cv2.face.FisherFaceRecognizer_create()
model.train(images,labels) 
face_cascade = cv2.CascadeClassifier(harr_file)

webcam = cv2.VideoCapture(0)
count = 0

while True:
    (_,img) = webcam.read()
    grayImg = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(grayImg , 1.3 , 5) # Detect faces in the grayscale frame

    for(x,y,w,h) in faces:
        face = grayImg[y:y+h , x:x+w]
        face_resize = cv2.resize(face , (width,height))
        prediction = model.predict(face_resize)  # Predict the label and confidence score for the face
        cv2.rectangle(img , (x,y),(x+w,y+h),(0,255,0),2)
        
        # Check if the confidence score indicates a known person
        if prediction[1] < 800:
            cv2.putText(img, '%s-%.0f' % (names[prediction[0]], prediction[1]), (x-10, y-10), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0))
            print(names[prediction[0]])
            count = 0
        else:
            count = 1
            cv2.putText(img , 'Unknown' , (x-10, y-10) , cv2.FONT_HERSHEY_PLAIN , 1 , (0,255,255))
             # If the counter exceeds 100 frames, save the image of the unknown person
            if (count > 100):
                print("Unkown Person")
                cv2.imwrite('input.jpg' , img)
                count = 0
                

    cv2.imshow('OpenCV' , img)
    key =cv2.waitKey(10)
    if key == 27:
        break

webcam.release()
cv2.destroyAllWindows()
```
