```python
import imutils
import cv2
```

```python
# Define HSV bounds for masking
Lower_bound = (86,53,64) 
Upper_bound = (174,192,255)
```

```python
cam = cv2.VideoCapture(0)

while True:
    (grabbed,frame) =cam.read()
    frame = imutils.resize(frame , width=900) #resize the image
    # Apply Gaussian Blur to smooth the image and reduce noise
    gaussian = cv2.GaussianBlur(frame,(11,11),0) # Kernel size is 11x11

    hsv = cv2.cvtColor(gaussian, cv2.COLOR_BGR2HSV) # change the color mode to HSV

    # Create a mask based on the defined HSV bounds
    mask = cv2.inRange(hsv,Lower_bound,Upper_bound)
    mask = cv2.erode(mask , None, iterations = 2) # remove small noise
    mask = cv2.dilate(mask , None , iterations = 2) # restore object size

    cnts = cv2.findContours(mask.copy() , cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
    center = None

    # If at least one contour is found
    if len(cnts) > 0 :
        c = max(cnts, key = cv2.contourArea) # Find the largest contour by area
        ((x,y),radius) = cv2.minEnclosingCircle(c) # Get the enclosing circle for the largest contour

        # Compute the center of the contour using image moments
        M = cv2.moments(c)
        center = (int(M["m10"]/M["m00"]) , int(M["m01"]/M["m00"]))

        # Draw the detected object and radius on the frame if radius is sufficient
        if radius > 10:
            cv2.circle(frame , (int(x),int(y)) , int(radius) , (0,255,0) , 2) # Draw the circle
            cv2.circle(frame , center , 5 , (0,255,0) , -1) # Mark the center point
            print("center :",center,"// radius :",radius)

            # Check proximity and object position for directional instructions
            if radius > 250 :
                print("Stop")
            else :
                if (center[0] < 150):
                    print("Right")
                elif (center[0] > 450):
                    print("Left")
                elif (radius < 250) :
                    print("Front")
                else :
                    print("Stop")
            
    
    cv2.imshow("Frame",frame)
    key = cv2.waitKey(1)
    if key == ord("q"):
        break

cam.release()
cv2.destroyAllWindows()
```
